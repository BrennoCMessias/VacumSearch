```python
import random

class VacuumEnvironment:
    def __init__(self, num_rooms, dirty_rooms):
        self.num_rooms = num_rooms
        self.dirty_rooms = dirty_rooms
        self.current_room = random.randint(0, num_rooms - 1)

    def is_dirty(self, room):
        return room in self.dirty_rooms

    def clean(self, room):
        self.dirty_rooms.remove(room)

    def move_left(self):
        if self.current_room > 0:
            self.current_room -= 1

    def move_right(self):
        if self.current_room < self.num_rooms - 1:
            self.current_room += 1

    def percept(self):
        return self.current_room, self.is_dirty(self.current_room)

def display_environment(environment):
    room_labels = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"]
    print("SALA       |", " | ".join(room_labels))
    print("Aspirador? |", " | ".join(["x" if room == environment.current_room else " " for room in range(environment.num_rooms)]))
    print("Sujeira    |", " | ".join(["x" if room in environment.dirty_rooms else " " for room in range(environment.num_rooms)]))
    print("\n")

def manual_control(environment):
    while environment.dirty_rooms:
        display_environment(environment)
        current_room, is_dirty = environment.percept()
        action = input("Escolha ação (Mover esquerda - 'E', Mover direita - 'D', Limpar - 'L', Esperar - 'ES'): ")
        if action == 'E':
            environment.move_left()
        elif action == 'D':
            environment.move_right()
        elif action == 'L':
            if is_dirty:
                environment.clean(current_room)
        elif action == 'ES':
            pass
        else:
            print("Ação inválida!")

    # Exibir ambiente após a limpeza
    display_environment(environment)

def base_control(environment):
    left_to_right = True
    while environment.dirty_rooms:
        display_environment(environment)
        current_room, is_dirty = environment.percept()
        if is_dirty:
            environment.clean(current_room)

        if left_to_right:
            if current_room < environment.num_rooms - 1:
                environment.move_right()
            else:
                left_to_right = False
        else:
            if current_room > 0:
                environment.move_left()
            else:
                left_to_right = True

    # Exibir ambiente após a limpeza
    display_environment(environment)

def oniscient_control(environment):
    while environment.dirty_rooms:
        display_environment(environment)
        dirtiest_room = max(range(environment.num_rooms), key=environment.is_dirty)
        while environment.current_room != dirtiest_room:
            if environment.current_room < dirtiest_room:
                environment.move_right()
            else:
                environment.move_left()
        if environment.is_dirty(dirtiest_room):
            environment.clean(dirtiest_room)

    # Exibir ambiente após a limpeza
    display_environment(environment)

def main():
    while True:
        while True:
            num_rooms = int(input("Número de salas (1 a 10): "))
            if 1 <= num_rooms <= 10:
                break
            else:
                print("Quantidade inválida. A quantidade de salas deve estar entre 1 e 10.")

        dirty_rooms = [i for i in range(num_rooms) if random.random() < 0.5]  # 50% chance of a room being dirty
        environment = VacuumEnvironment(num_rooms, dirty_rooms)

        print("\nAmbiente Inicial:")
        display_environment(environment)

        print("\nEscolha o controlador:")
        print("1 - Manual")
        print("2 - Base")
        print("3 - Onisciente")

        choice = input("Escolha o controlador (1/2/3): ")

        if choice == '1':
            manual_control(environment)
        elif choice == '2':
            base_control(environment)
        elif choice == '3':
            oniscient_control(environment)
        else:
            print("Escolha inválida!")

        restart = input("\nDeseja reiniciar o programa (S para sim, qualquer outra tecla para sair)? ")
        if restart.upper() != 'S':
            break

if __name__ == "__main__":
    main()

#IMPLEMENTAR A IPÇÃO 1, MAS SEM VER A TABELA COMPLETA
#completo abaixo
```

```python
import random

class VacuumEnvironment:
    def __init__(self, num_rooms, dirty_rooms):
        self.num_rooms = num_rooms
        self.dirty_rooms = dirty_rooms
        self.current_room = random.randint(0, num_rooms - 1)

    def is_dirty(self, room):
        return room in self.dirty_rooms

    def clean(self, room):
        self.dirty_rooms.remove(room)

    def move_left(self):
        if self.current_room > 0:
            self.current_room -= 1

    def move_right(self):
        if self.current_room < self.num_rooms - 1:
            self.current_room += 1

    def percept(self):
        return self.current_room, self.is_dirty(self.current_room)

def display_environment(environment):
    room_labels = ["A", "B", "C", "D", "E", "F", "G", "H", "I", "J"]
    print("SALA       |", " | ".join(room_labels))
    print("Aspirador? |", " | ".join(["x" if room == environment.current_room else " " for room in range(environment.num_rooms)]))
    print("Sujeira    |", " | ".join(["x" if room in environment.dirty_rooms else " " for room in range(environment.num_rooms)]))
    print("\n")

def manual_control(environment):
    show_environment = input("Deseja exibir o ambiente a cada ação? (S para sim, qualquer outra tecla para não): ").upper() == 'S'

    while environment.dirty_rooms:
        if show_environment:
            display_environment(environment)

        current_room, is_dirty = environment.percept()
        action = input("Escolha ação (Mover esquerda - 'E', Mover direita - 'D', Limpar - 'L', Esperar - 'ES'): ")

        if action == 'E':
            environment.move_left()
        elif action == 'D':
            environment.move_right()
        elif action == 'L':
            if is_dirty:
                environment.clean(current_room)
        elif action == 'ES':
            pass
        else:
            print("Ação inválida!")

    # Exibir ambiente após a limpeza
    display_environment(environment)

def base_control(environment):
    left_to_right = True
    while environment.dirty_rooms:
        display_environment(environment)
        current_room, is_dirty = environment.percept()
        if is_dirty:
            environment.clean(current_room)

        if left_to_right:
            if current_room < environment.num_rooms - 1:
                environment.move_right()
            else:
                left_to_right = False
        else:
            if current_room > 0:
                environment.move_left()
            else:
                left_to_right = True

    # Exibir ambiente após a limpeza
    display_environment(environment)

def oniscient_control(environment):
    while environment.dirty_rooms:
        display_environment(environment)
        dirtiest_room = max(range(environment.num_rooms), key=environment.is_dirty)
        while environment.current_room != dirtiest_room:
            if environment.current_room < dirtiest_room:
                environment.move_right()
            else:
                environment.move_left()
        if environment.is_dirty(dirtiest_room):
            environment.clean(dirtiest_room)

    # Exibir ambiente após a limpeza
    display_environment(environment)

def main():
    while True:
        while True:
            num_rooms = int(input("Número de salas (1 a 10): "))
            if 1 <= num_rooms <= 10:
                break
            else:
                print("Quantidade inválida. A quantidade de salas deve estar entre 1 e 10.")

        dirty_rooms = [i for i in range(num_rooms) if random.random() < 0.5]  # 50% chance of a room being dirty
        environment = VacuumEnvironment(num_rooms, dirty_rooms)

        print("\nAmbiente Inicial:")
        display_environment(environment)

        initial_state = environment.current_room, environment.dirty_rooms.copy()

        print("\nEscolha o controlador:")
        print("1 - Manual")
        print("2 - Base")
        print("3 - Onisciente")

        choice = input("Escolha o controlador (1/2/3): ")

        if choice == '1':
            manual_control(environment)
        elif choice == '2':
            base_control(environment)
        elif choice == '3':
            oniscient_control(environment)
        else:
            print("Escolha inválida!")

        final_state = environment.current_room, environment.dirty_rooms.copy()

        print("\nEstado do Ambiente Antes:")
        environment.current_room, environment.dirty_rooms = initial_state
        display_environment(environment)

        print("\nEstado do Ambiente Depois:")
        environment.current_room, environment.dirty_rooms = final_state
        display_environment(environment)

        restart = input("\nDeseja reiniciar o programa (S para sim, qualquer outra tecla para sair)? ")
        if restart.upper() != 'S':
            break

if __name__ == "__main__":
    main()
#é possível escolher se vc quer exibir o ambiente ou não
```
